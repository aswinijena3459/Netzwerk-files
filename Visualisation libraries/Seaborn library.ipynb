### relplot()- to understand the relationships between the variables of a dataset

```python
sns.set(style='darkgrid')
sns.relplot(x='passengers',y='year',data=df)
```

#### no of passengers kept increasing gradually

```python
sns.relplot(x='passengers',y='year',hue='year',data=df)
```

### lineplot- can be use by using kind='line'

```python
sns.relplot(x='passengers',y='year',hue='year',data=df,kind='line')
```

### Plotting with Categorical Data:
This approach comes into the picture when our main variable is further divided into discrete groups (categorical). This can be achieved using the catplot() function.

```python
df=sns.load_dataset('tips')
df
```

```python
sns.catplot(x='day',y='total_bill',data=df)
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm')
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='point')
```

### Multi-Plot Grids :- plots or graphs that are plotted using the same scale and axes to aid comparison between them

```python
a=sns.load_dataset('iris')
a
```

```python
b = sns.FacetGrid(a, col="species")
b.map(plt.scatter, "sepal_length",'petal_length')  #Apply a plotting function to each facet's subset of the data.
```

```python
b.map(plt.hist, "sepal_length")
```
