```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,20)
y=np.arange(30,50)
plt.scatter(x,y,c='b',marker='*')
plt.xlabel=("x-axis")
plt.ylabel=('Y-axis')
plt.title=('scatter plot')
plt.show()
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,'r')
plt.xlabel='x axis'
plt.ytitle='y axis'
plt.title='2d graph'
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y= np.sin(x)
plt.plot(x,y,'r')
plt.xlabel=('x axis')
plt.ylabel=('y axis')
plt.title=('sine wave')
plt.show()
```

```python
print(x)
print(y)
```

```python
x1=np.arange(0,2*np.pi,0.1)
y1=np.cos(x1)

plt.plot(x1,y1,"g")
plt.xlabel='x axis'
plt.ylabel='y axis'
plt.title='cosine wave'
plt.show()
```

```python
standard=['10th','12th','nursery']
strength=[45,78,56]

plt.bar(standard,strength)
plt.xlabel='standard'
plt.ylabel='strength'
plt.title=('whatever')
plt.show()
```

```python
x = np.array([22,87,5,43,56,51,5,79,31,27]) 
plt.hist(x)
plt.xlabel='x axis'
plt.ylabel='y axis'
plt.title='histogram'
plt.show()
```

```python
x = np.array([22,87,5,43,56,8,73,55,54,11,20,51,5,79,31,27]) 

plt.hist(x,bins=20)
plt.xlabel='x axis'
plt.ylabel='y axis'
plt.xticks=np.arange(0,90,10)
plt.title='diff'
plt.show()
```
