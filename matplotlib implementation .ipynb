```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1: Read Total profit of all months and show it using a line plot

```python
df=pd.read_csv(r"C:\Users\Aswini\Downloads\company_sales_data.csv")
```

```python
df
```

```python
month=df['month_number'].tolist()
profit=df['total_profit'].tolist()
plt.plot(month,profit)
plt.xlabel('month')
plt.xticks(np.arange(1,13))
plt.ylabel('profits')
plt.yticks(np.arange(100000,600000,100000))
plt.title('profit per month data')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –

Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

Line width should be 3

```python
month=df['month_number'].tolist()
profit=df['total_profit'].tolist()
plt.plot(month,profit,label='profit data',linestyle='--',linewidth=3,color='r',marker='o')
plt.xlabel('month')

plt.legend(loc='lower right')
plt.xticks(np.arange(1,13))
plt.ylabel('profits')
plt.yticks(np.arange(100000,600000,100000))
plt.title('profit per month data')
plt.show()
```

### 3: Read all product sales data and show it  using a multiline plot

```python
df
```

```python
month=df['month_number'].tolist()
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
toothpaste=df['toothpaste'].tolist()
bathingsoap=df['bathingsoap'].tolist()
shampoo=df['shampoo'].tolist()
moisturizer=df['moisturizer'].tolist()


plt.plot(month,facecream,label='facecream sales data',linewidth=3,marker='o',)
plt.plot(month,facewash,label='facewash sales data',linewidth=3,marker='o',)
plt.plot(month,toothpaste,label='toothpaste sales data',linewidth=3,marker='o',)
plt.plot(month,bathingsoap,label='bathingsoap sales data',linewidth=3,marker='o',)
plt.plot(month,shampoo,label='shampoo sales data',linewidth=3,marker='o',)
plt.plot(month,moisturizer,label='moisturizer sales data',linewidth=3,marker='o',)

plt.legend(loc='upper left')
plt.xlabel('month')
plt.xticks(np.arange(1,13))
plt.ylabel("sales")
plt.yticks(np.arange(1000,16000,1000))

plt.show()

```

### 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
toothpaste=df['toothpaste'].tolist()
month=df['month_number'].tolist()

plt.scatter(month,toothpaste,label='toothpaste data')
plt.xlabel("months")
plt.xticks(np.arange(1,13))
plt.ylabel('no of units sold')
plt.yticks(np.arange(3000,12000,1000))
plt.title('toothpaste sales data')
plt.legend(loc='upper left')
plt.show()


```

### 5: Read face cream and facewash product sales data and show it using the bar chart

```python
month=df['month_number'].tolist()
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()

plt.bar([a-0.25 for a in month],facecream,align='edge',width= 0.25,label = 'Face Cream sales data')
plt.bar([a+0.25 for a in month],facewash,align='edge',width= -0.25,label = 'Face wash sales data')
plt.xlabel('months')
plt.xticks(np.arange(1,13))
plt.ylabel('sales')
plt.yticks(np.arange(200,4000,200))
plt.legend(loc='upper left')
plt.title('mothly sales figures')
plt.show()
```

### 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
totalprofit=df['total_profit'].tolist()
plt.hist(totalprofit,label='max profit ranges')
plt.title('common profit ranges')
plt.xlabel('profit ranges')
plt.xticks(np.arange(150000,500000,50000))
plt.ylabel('range in dollars')
plt.legend(loc='upper left')
plt.show()
```

### 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
months=df['month_number'].tolist()
facewash=df['facewash'].tolist()
bathingsoap=df['bathingsoap'].tolist()

f,axarr=plt.subplots(2,sharex=True)
axarr[0].plot(months,facewash,label='facewash sales',linewidth=3,color='k',marker='o')
axarr[0].set_title('sales data of face wash')
axarr[1].plot(months,bathingsoap,label='bathing soap sales',linewidth=3,color='r',marker='o')
axarr[1].set_title('sales data of bathing soap')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```
