### handling missing values

```python
# trying to handle missing values of categorical features
```

```python
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## Replacinf missing value with a new label
def replace_nan_feature(dataset,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df=replace_nan_feature(df,features_nan)

df[features_nan].isnull().sum()
```

```python
## so that now we can creatr a new column also taking 0 and 1 as values 
### replacing null values with a new label can be used to handle missing values in categorical features
```

### missing values in numerical features

```python
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in numerical_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## we will repalce the nana values with median because during eda we saw there are a lot of outliers oin the 
#numerical features
```

```python
for feature in numerical_nan:
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'_nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_nan].isnull().sum()
```
