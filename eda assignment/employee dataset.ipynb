```python
sns.catplot(x='left',y='satisfaction_level',data=df,kind='bar')
```

#### those employee who were more satisfied didnot leave the company

```python
sns.distplot(df['satisfaction_level'])
```

#### very few employees are dissatisfied with the company

```python
sns.jointplot(x='satisfaction_level',y='average_montly_hours',data=df,kind='reg')
```

#### employees who are satisfied with the company work for 10hrs a day on an average

```python
sns.catplot(x='time_spend_company',y='satisfaction_level',data=df)
```

```python
sns.distplot(df['time_spend_company'])
```

#### most employees work for generally 2 - 4 years in the company on an average

```python
sns.boxplot(x='time_spend_company',y='satisfaction_level',data=df)
```

#### few people worked for 2 years inspite of dissatisfaction with the company ,may be due some financial problems
